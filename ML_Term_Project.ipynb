## KNN Classifier

```python
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('preprocessed_data.csv')
X = df.drop(['Accept'], axis = 1)
y = df[['Accept']]


# One-hot encoding for categorical data and labels
X_encoded_data = pd.get_dummies(X, columns = ['A1', 'A2','A3','A4','A5','A6','A7','A8','A9'])
y_encoded_data = pd.get_dummies(y, columns = ['Accept'])

# Splitting into train and test data
X_train, X_test, y_train, y_test = train_test_split(
        X_encoded_data, y_encoded_data, test_size=0.2, random_state=42
    )
```

```python
# DataFrame for storing results
df_knn = pd.DataFrame(columns=['k', 'Accuracy'])
```

```python
# Iterating over odd values of k 
for k in range(1,21,2):
  Classifier = KNeighborsClassifier(n_neighbors= k)
  Classifier.fit(X_train, y_train)
  predictions = Classifier.predict(X_test)
  df_knn.loc[len(df_knn.index)] = [k, accuracy_score(y_test,predictions)] 
```

```python
df_knn
```

## Distance weighted KNN

```python
# DataFrame for storing results
df_knn_dist = pd.DataFrame(columns=['k', 'Accuracy'])
```

```python
# Iterating over odd values of k 
for i in range(1,21,2):
  Classifier = KNeighborsClassifier(n_neighbors= i, weights='distance')
  Classifier.fit(X_train,y_train)
  predictions = Classifier.predict(X_test)
  df_knn_dist.loc[len(df_knn_dist.index)] = [i, accuracy_score(y_test,predictions)] 
```

```python
df_knn_dist
```

## Naive Bayes

```python
from sklearn.naive_bayes import CategoricalNB
import numpy as np
from sklearn.metrics import accuracy_score
```

```python
# Changing y_train and y_test to expected input format for sklearn
y_train = (y_train['Accept_+']==1).astype(int)
y_test = (y_test['Accept_+']==1).astype(int)
```

```python
# Fit the Naive Bayes classifier on training data
clf = CategoricalNB()
clf.fit(X_train, y_train)

# Making predictions on test data
test_predictions = clf.predict(X_test)

# Making predictions on training data
train_predictions = clf.predict(X_train)
```

```python
# DataFrame for storing results
df_nb = pd.DataFrame(columns=['Dataset', 'Accuracy'])
df_nb.loc[0] = ['Training Data', accuracy_score(y_train, train_predictions)] 
df_nb.loc[1] = ['Testing Data', accuracy_score(y_test, test_predictions)]
```
